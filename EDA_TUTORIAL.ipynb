### 3. Check for relationship/patterns in the dataset

### A. Correlation Coefficient
which is classified into:
a. Heatmap(heatmap)- indicates the relationship between the attributes of the data
b. Pairplot(pairplot)- indicates the relationship between only 2 attributes of the data
c. Scatterplot(relplot) - indicates the relationship between 3 attributes of the data
d. Histogram(distplot) - indicates an attribute with the range of values and occurence
e. (catplot)- distinquishes the value of an attribute with the highest occurence using shapes like box,

```python
correlation = DiabetesNew.corr() # to check for correlation in the dataset
sns.heatmap(correlation, xticklabels =correlation.columns , yticklabels =correlation.columns , annot = True) 
#sns is a library, heatmap is a function in the library, and the content in the parenthesis are referred to as parameters
#annot(annotation, which indicates the correlation value of each attributes)
```

```python
sns.heatmap(correlation, xticklabels = correlation.columns, yticklabels = correlation.columns, annot = False)
```

```python
sns.heatmap(correlation, xticklabels = correlation.columns, yticklabels = correlation.columns)
#by default, annot is set to false
```

```python
sns.pairplot(DiabetesNew)
```

```python
sns.relplot(x ="BloodPressure", y ="Insulin", hue ="Outcome", data = DiabetesNew)
#indicates the relationship between 3 attributes in the dataset. relplot is the module(function) name for scatterplot.
#x = x axis, y = y axis and hue = the relationship between the 3 attributes
```

```python
sns.distplot(DiabetesNew["BMI"], bins = 10) #histogram to check the range of values for an attribute and their occurence
```

```python
sns.catplot(x = "BMI", kind = "box", data = DiabetesNew) 
#catplot is a module in seaborn(sns) that is used to indicate the highest occuring value in an attribute. 
#x = attribute(column) of interest, kind = shape type, data = name of dataset
```
